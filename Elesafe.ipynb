```python
import zipfile
import os

zip_file = "Object Detection.v1i.yolov8.zip"
data_dir = "Object Detection_dataset"

# Unzip
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(data_dir)

# Check the folder structure
os.listdir(data_dir)
```

```python
# Example path to data.yaml
data_yaml_path = os.path.join(data_dir, "data.yaml")
print(data_yaml_path)
```

```python
from ultralytics import YOLO

# Create a YOLO model (you can use "yolov8n.pt" for nano or "yolov8s.pt" for small)
model = YOLO("yolov8n.pt")

# Train the model
model.train(
    data=data_yaml_path,
    epochs=50,          # You can change epochs
    imgsz=640,          # Image size
    batch=16,           # Batch size
    name="yolov8_custom" # Save results in this folder
)
```

```python
# Test on validation or test set
results = model.val(data=data_yaml_path)

# OR predict on a single image
results = model.predict(source=os.path.join(data_dir, "test/images"))
results.show()
```
